# 1d.	Hacer el mapa de calor de contagios (o infectados) o muertes(o fallecidos) o  por departamentos o por distritos de Lima.

```python
#importamos las librerias necesarias

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```


# *Archivo de los "FALLECIDOS y POSITIVOS" por DEPARTAMENETOS*

```python
#llamamos al doc correspondiente de los casos de fallecidos POR DEPARTAMENTO
dfn = pd.read_excel('cofallecidos.xlsx', sheet_name = 'DEPA')
#IMPRIMIMOS LA TABLA
dfn.head()
```

### Mapa de calor de los "FALLECIDOS" por DEPARTAMENTO

```python
#AQUI UBICAMOS LAS VARIABLES CORRESPONDIENTES EN EL CUADRO DE DOBLE ENTRADA 
datos = dfn.pivot(index="DEPARTAMENTO", columns='MESES',values="FALLECIDOS")
datos
```

```python
#AQUI DIBUJAMOS LA MAPA DE CALOR DE FALLECIDOS POR DEPARTAMENTO EN LOS MESES DE MARZO HASTA SEPTIEMBRE
plt.figure(figsize=(12,12))#ASIGNAMOS LAS MEDIDAS DE LA GRAFICA, PARA UNA MEJOR VISUALIZACION 
sns.heatmap(data=datos, annot=True, fmt='d', cmap='YlGn')#con annot: le asignamos los datos, con fmt: los datos seran enteros y con cmap: le asignamos un color 
plt.show()
```

## Mapa de calor de los "POSITIVOS" por DEPARTAMENTO 

```python
#AQUI UBICAMOS LAS VARIABLES CORRESPONDIENTES EN EL CUADRO DE DOBLE ENTRADA 
datos1 = dfn.pivot(index="DEPARTAMENTO", columns='MESES',values='POSITIVOS')
datos1 #mostramos los datos 
```

```python
#AQUI DIBUJAMOS LA MAPA DE CALOR DE POSITIVOS POR DEPARTAMENTO EN LOS MESES DESDE MARZO HASTA SEPTIEMBRE 
plt.figure(figsize=(13,13))#ASIGNAMOS LAS MEDIDAS DE LA GRAFICA, PARA UNA MEJOR VISUALIZACION 
sns.heatmap(data=datos1, annot=True, fmt='d', cmap='PuRd')#con annot: le asignamos los datos, con fmt: los datos seran enteros y con cmap: le asignamos un color 
plt.show()
```

# Informacion de los "FALLECIDOS y POSITIVOS" en LIMA

```python
#llamamos al doc correspondiente de los casos de fallecidos en LIMA
dflima = pd.read_excel('cofallecidos.xlsx', sheet_name = 'LIMA')
#IMPRIMIMOS LA TABLA
dflima.head()
```

### Aqui veremos el mapa de calor de los "FALLECIDOS" en LIMA

```python
#AQUI UBICAMOS LAS VARIABLES CORRESPONDIENTES EN EL CUADRO DE DOBLE ENTRADA 
datoslima = dflima.pivot(index="DISTRITO - LIMA", columns='MESES',values="LIMA - FALLECIDOS")
datoslima #presentamos la tabla de datos 
```

```python
#AQUI DIBUJAMOS LA MAPA DE CALOR DE FALLECIDOS POR DEPARTAMENTO EN LOS MESES DE MARZO HASTA SEPTIEMBRE
plt.figure(figsize=(12,18))#ASIGNAMOS LAS MEDIDAS DE LA GRAFICA, PARA UNA MEJOR VISUALIZACION 
sns.heatmap(data=datoslima, cmap='YlOrBr')# con cmap: le asignamos un color 
plt.show()
```

### Aqui veremos el mapa de calor de los "POSITIVOS" en LIMA

```python
#AQUI UBICAMOS LAS VARIABLES CORRESPONDIENTES EN EL CUADRO DE DOBLE ENTRADA 
datoslima1 = dflima.pivot(index="DISTRITO - LIMA", columns='MESES',values="POSITIVOS")
datoslima1 #presentamos la tabla de datos
```

```python
#AQUI DIBUJAMOS LA MAPA DE CALOR DE FALLECIDOS POR DEPARTAMENTO EN LOS MESES DE MARZO HASTA SEPTIEMBRE
plt.figure(figsize=(12,20))#ASIGNAMOS LAS MEDIDAS DE LA GRAFICA, PARA UNA MEJOR VISUALIZACION 
sns.heatmap(data=datoslima1, annot=True, fmt='d',cmap='gist_rainbow')#con cmap: le asignamos un color 
plt.show()
```
